6)What are Type 1 and Type 2 errors in hypothesis testing?

Ans:-Type I and Type II Errors in Hypothesis Testing
When making a decision in hypothesis testing, two types of errors can occur:

1. Type I Error (False Positive)
Occurs when we reject the null hypothesis (H‚ÇÄ) when it is actually true.
It is also called a false positive because we detect an effect that does not actually exist.
The probability of making a Type I error is the significance level (Œ±), commonly set at 0.05 (5%).
Example: A test incorrectly concludes that a new drug is effective when it actually has no effect.

2. Type II Error (False Negative)
Occurs when we fail to reject H‚ÇÄ when the alternative hypothesis (H‚ÇÅ) is actually true.
It is also called a false negative because we fail to detect a real effect.
The probability of making a Type II error is Œ≤, and 1 - Œ≤ is the statistical power of the test.
Example: A test fails to detect that a new drug is effective when it actually works.

20)What is an ANOVA test, and what are its assumptions?

Ans:-
ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if there is a statistically significant difference between them. It helps in identifying whether at least one group mean is different from the others without conducting multiple t-tests, which could increase the risk of Type I errors.

### *Assumptions of ANOVA*
For ANOVA results to be valid, the following assumptions must be met:

1. **Independence**: The observations must be independent of each other. This means that data points in one group should not influence those in another.

2. **Normality**: The data in each group should be approximately normally distributed. This assumption is especially important for small sample sizes but can be relaxed for larger samples (Central Limit Theorem).

3. **Homogeneity of Variance (Homoscedasticity)**: The variance within each group should be roughly equal. This is tested using Levene‚Äôs test or Bartlett‚Äôs test.

4. **Random Sampling**: The data should be collected using a random sampling method to ensure generalizability.

If these assumptions are violated, alternative approaches such as **Welch‚Äôs ANOVA** (for unequal variances) or **non-parametric tests** (like Kruskal-Wallis test) can be considered.

21)What are the different types of ANOVA tests?

Ans:-1. One-way ANOVA

Purpose: Used to compare the means of three or more independent groups on a single dependent variable.

Factors: Two independent variables (e.g., teaching method and gender).

Purpose: An extension of two-way ANOVA, used when there are three or more independent variables (factors) to examine their individual and combined effects on a dependent variable.

Factors: Three or more independent variables.

Purpose: Used when the same group of subjects is measured multiple times under different conditions or time points. This allows for analyzing changes within subjects.
Example: Assessing the effect of a drug on blood pressure by measuring it before, during, and after treatment in the same individuals.
Within-subjects factor: The factor that represents the different conditions or time points (e.g., time of measurement).

Purpose: Combines elements of between-subjects and within-subjects designs. It involves at least one between-subjects factor (different groups) and one within-subjects factor (repeated measures).
Example: Comparing the effectiveness of two different therapies (between-subjects) on anxiety levels measured at multiple time points (within-subjects).

Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
    """
    Performs a one-sample Z-test.

    Args:
        sample_mean: The mean of the sample.
        pop_mean: The known population mean.
        pop_std: The known population standard deviation.
        sample_size: The size of the sample.
        alpha: The significance level (default 0.05).

    Returns:
        z_score: The calculated Z-score.
        p_value: The two-tailed p-value.
        interpretation: A string interpreting the results.
    """

    # Calculate the Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Calculate the two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    # Interpret the results
    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than alpha ({alpha}). " \
                       f"We reject the null hypothesis. There is statistically significant " \
                       f"evidence that the sample mean is different from the population mean."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than or equal to alpha ({alpha}). " \
                       f"We fail to reject the null hypothesis. There is not enough " \
                       f"evidence to suggest that the sample mean is different from the population mean."

    return z_score, p_value, interpretation


# Example usage:
sample_mean = 155  # Sample mean of heights (cm)
pop_mean = 160    # Known population mean of heights (cm)
pop_std = 10      # Known population standard deviation of heights (cm)
sample_size = 50   # Sample size

z_score, p_value, interpretation = z_test(sample_mean, pop_mean, pop_std, sample_size)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.3f}")
print(interpretation)


# Example 2 (Illustrating a different outcome)
sample_mean = 162  # Sample mean of heights (cm)
pop_mean = 160    # Known population mean of heights (cm)
pop_std = 10      # Known population standard deviation of heights (cm)
sample_size = 50   # Sample size

z_score, p_value, interpretation = z_test(sample_mean, pop_mean, pop_std, sample_size)

print(f"\nExample 2:")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.3f}")
print(interpretation)
```

Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# Simulated sample data (normally distributed)
sample = np.random.normal(loc=100, scale=15, size=30)  # Sample with mean 100, std dev 15, size 30

# Known population parameters
pop_mean = 105  # Hypothesized population mean
pop_std = 15    # Known population standard deviation

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Compute the Z-score
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

# Compute the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis testing interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulated sample data (normally distributed)
sample = np.random.normal(loc=100, scale=15, size=30)  # Sample with mean 100, std dev 15, size 30

# Known population parameters
pop_mean = 105  # Hypothesized population mean
pop_std = 15    # Known population standard deviation

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Compute the Z-score
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

# Compute the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis testing interpretation
alpha = 0.05  # Significance level
z_critical = stats.norm.ppf(1 - alpha / 2)  # Two-tailed test (1.96 for 95% confidence)
reject_region = [-z_critical, z_critical]

if abs(z_score) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")

# Visualization of decision region
x = np.linspace(-4, 4, 1000)  # Z-score range
y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Highlight rejection regions
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label="Rejection Region")

# Plot critical z-scores
plt.axvline(-z_critical, color='red', linestyle='--', label=f"Critical Z = {-z_critical:.2f}")
plt.axvline(z_critical, color='red', linestyle='--', label=f"Critical Z = {z_critical:.2f}")

# Plot computed Z-score
plt.axvline(z_score, color='black', linestyle='-', label=f"Z-score = {z_score:.2f}")

# Labels and legend
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-Test Decision Region")
plt.legend()
plt.show()
```

 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type_errors(mu0=100, mu_alt=105, sigma=15, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors in hypothesis testing.

    Parameters:
    - mu0: Population mean under null hypothesis (H0)
    - mu_alt: Alternative mean (H1)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default: 0.05)
    """

    # Standard error of the mean
    se = sigma / np.sqrt(n)

    # Critical Z-score for a one-tailed test (right-tailed)
    z_critical = stats.norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se  # Convert Z-score to raw value

    # Define X-axis range
    x = np.linspace(mu0 - 4 * se, mu_alt + 4 * se, 1000)

    # Null hypothesis distribution (H0: true mean is mu0)
    y_h0 = stats.norm.pdf(x, mu0, se)

    # Alternative hypothesis distribution (H1: true mean is mu_alt)
    y_h1 = stats.norm.pdf(x, mu_alt, se)

    # Compute beta (Type II error probability)
    beta = stats.norm.cdf(x_critical, mu_alt, se)
    power = 1 - beta  # Power of the test

    # Plot distributions
    plt.figure(figsize=(9, 5))
    plt.plot(x, y_h0, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, y_h1, label="Alternative Hypothesis (H1)", color="green")

    # Type I Error Region (Alpha)
    plt.fill_between(x, y_h0, where=(x >= x_critical), color='red', alpha=0.3, label="Type I Error (Œ±)")

    # Type II Error Region (Beta)
    plt.fill_between(x, y_h1, where=(x <= x_critical), color='orange', alpha=0.3, label="Type II Error (Œ≤)")

    # Critical value line
    plt.axvline(x_critical, color='black', linestyle='--', label=f"Critical Value = {x_critical:.2f}")

    # Labels and legend
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.legend()
    plt.show()

    # Print error probabilities
    print(f"Type I Error (Œ±): {alpha:.2f}")
    print(f"Type II Error (Œ≤): {beta:.2f}")
    print(f"Power of the test (1 - Œ≤): {power:.2f}")

# Run the function with default values
visualize_type_errors()
```

Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - sample: List or NumPy array of sample data.
    - confidence: Confidence level (default is 95%).

    Returns:
    - (lower_bound, upper_bound): Confidence interval for the mean.
    """

    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)
    sample_size = len(sample)

    # Compute the critical t-value (since population std is unknown)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    # Compute margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample
ci = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculates the margin of error for a given sample and confidence level.

    Parameters:
    - sample: List or NumPy array of sample data.
    - confidence: Confidence level (default is 95%).

    Returns:
    - Margin of error (ME)
    """

    # Sample statistics
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)

    # Compute the critical t-value (for small sample size, we use t-distribution)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    # Compute margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    return margin_of_error

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample
me = margin_of_error(sample_data, confidence=0.95)

print(f"Margin of Error (95% Confidence Level): {me:.2f}")
```

 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to perform Bayesian Inference for a Coin Flip Problem
def bayesian_inference(prior_heads=0.5, prior_tails=0.5, flips=10, observed_heads=7):
    """
    Perform Bayesian Inference for estimating a biased coin.

    Parameters:
    - prior_heads: Prior probability of heads (fair coin, start with 50%).
    - prior_tails: Prior probability of tails.
    - flips: Total number of coin flips.
    - observed_heads: Number of heads observed in flips.

    Returns:
    - Posterior probability of heads.
    """

    # Likelihood of observing the data given the hypothesis (binomial distribution)
    likelihood = (prior_heads ** observed_heads) * (prior_tails ** (flips - observed_heads))

    # Prior belief (using uniform distribution for simplicity, representing fair coin)
    prior = prior_heads

    # Evidence (Marginal likelihood): normalizing constant
    evidence = prior_heads * (prior_heads ** observed_heads) * (prior_tails ** (flips - observed_heads)) + \
               prior_tails * (prior_heads ** observed_heads) * (prior_tails ** (flips - observed_heads))

    # Posterior Probability: Applying Bayes' Theorem
    posterior = (likelihood * prior) / evidence

    return posterior

# Example with 10 coin flips and 7 heads observed
posterior_probability = bayesian_inference(prior_heads=0.5, prior_tails=0.5, flips=10, observed_heads=7)

# Print the posterior probability
print(f"Posterior probability of heads after observing 7 heads in 10 flips: {posterior_probability:.4f}")

# Visualization of Posterior Distribution (with a prior belief)
# Let's plot posterior probability vs the number of heads observed
x = np.arange(0, 11)
y = [bayesian_inference(prior_heads=0.5, prior_tails=0.5, flips=10, observed_heads=i) for i in x]

# Plotting the posterior probability distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='b', label="Posterior Probability of Heads")
plt.xlabel("Number of Heads Observed")
plt.ylabel("Posterior Probability")
plt.title("Bayesian Inference: Posterior Probability vs. Observed Heads")
plt.grid(True)
plt.legend()
plt.show()
```

Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed_frequencies, expected_probabilities, alpha=0.05):
    """
    Performs a Chi-square goodness-of-fit test.

    Args:
        observed_frequencies: A list or numpy array of observed frequencies for each category.
        expected_probabilities: A list or numpy array of expected probabilities for each category.
                                 These should sum to 1.
        alpha: The significance level (default 0.05).

    Returns:
        chi2_statistic: The calculated Chi-square statistic.
        p_value: The p-value.
        degrees_of_freedom: The degrees of freedom.
        interpretation: A string interpreting the results.
    """

    observed_frequencies = np.array(observed_frequencies)  # Convert to NumPy array for calculations
    expected_frequencies = np.array(expected_probabilities) * np.sum(observed_frequencies) #calculate expected frequencies

    degrees_of_freedom = len(observed_frequencies) - 1  # Number of categories - 1

    chi2_statistic, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)  # Chi-square test

    interpretation = interpret_results(p_value, alpha)

    return chi2_statistic, p_value, degrees_of_freedom, interpretation


def interpret_results(p_value, alpha):
    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than alpha ({alpha}). " \
                       f"We reject the null hypothesis. The observed distribution is significantly " \
                       f"different from the expected distribution."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than or equal to alpha ({alpha}). " \
                       f"We fail to reject the null hypothesis. There is not enough evidence to " \
                       f"suggest that the observed distribution is different from the expected distribution."
    return interpretation


# Example 1: Testing if a die is fair
observed_frequencies = [12, 8, 11, 9, 10, 10]  # Observed frequencies for each face of a die
expected_probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]  # Expected probabilities for a fair die

chi2_statistic, p_value, degrees_of_freedom, interpretation = goodness_of_fit_test(observed_frequencies, expected_probabilities)

print("Goodness-of-Fit Test (Die Example):")
print(f"Chi-square statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(interpretation)


# Example 2: Testing if coin flips follow a binomial distribution
observed_frequencies = [55, 45]  # 55 heads, 45 tails in 100 flips
expected_probabilities = [0.5, 0.5]  # Expected probabilities for a fair coin

chi2_statistic, p_value, degrees_of_freedom, interpretation = goodness_of_fit_test(observed_frequencies, expected_probabilities)

print("\nGoodness-of-Fit Test (Coin Flips):")
print(f"Chi-square statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(interpretation)


# Example 3: Unequal probabilities
observed_frequencies = [30, 20, 50]
expected_probabilities = [0.2, 0.2, 0.6]  # Example of unequal probabilities. Must sum to 1.

chi2_statistic, p_value, degrees_of_freedom, interpretation = goodness_of_fit_test(observed_frequencies, expected_probabilities)

print("\nGoodness-of-Fit Test (Unequal probabilities):")
print(f"Chi-square statistic: {chi2_statistic:.2f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(interpretation)
```

 Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
import scipy.stats as stats

# Simulating two random samples
np.random.seed(0)  # For reproducibility

# Sample 1: Data from a normal distribution with mean=0 and std=1
sample_1 = np.random.normal(loc=0, scale=1, size=50)

# Sample 2: Data from a normal distribution with mean=0 and std=2 (larger variance)
sample_2 = np.random.normal(loc=0, scale=2, size=50)

# Step 1: Calculate sample variances
var_1 = np.var(sample_1, ddof=1)  # Sample variance for sample 1 (using ddof=1 for sample variance)
var_2 = np.var(sample_2, ddof=1)  # Sample variance for sample 2

# Step 2: Calculate the F-statistic
F_statistic = var_1 / var_2 if var_1 > var_2 else var_2 / var_1  # F-statistic is the ratio of variances

# Step 3: Calculate degrees of freedom for both samples
df_1 = len(sample_1) - 1  # Degrees of freedom for sample 1
df_2 = len(sample_2) - 1  # Degrees of freedom for sample 2

# Step 4: Perform the F-test (Calculate the p-value)
p_value = 1 - stats.f.cdf(F_statistic, df_1, df_2)  # Using the CDF of the F-distribution

# Print results
print(f"Sample 1 Variance: {var_1:.4f}")
print(f"Sample 2 Variance: {var_2:.4f}")
print(f"F-statistic: {F_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

# Hypothesis Testing
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for three groups
np.random.seed(0)  # For reproducibility

# Group 1: Data with mean=50 and std=10
group_1 = np.random.normal(loc=50, scale=10, size=30)

# Group 2: Data with mean=55 and std=10
group_2 = np.random.normal(loc=55, scale=10, size=30)

# Group 3: Data with mean=60 and std=10
group_3 = np.random.normal(loc=60, scale=10, size=30)

# Step 1: Perform the one-way ANOVA
f_statistic, p_value = stats.f_oneway(group_1, group_2, group_3)

# Print the results
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 2: Hypothesis Testing
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is significantly different.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between group means.")

# Step 3: Plotting the results - Boxplot to visualize group distributions
plt.figure(figsize=(8, 6))
sns.boxplot(data=[group_1, group_2, group_3], notch=True, patch_artist=True)
plt.title('One-Way ANOVA Test: Group Means Comparison')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.grid(True)
plt.show()
```

 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Function to check the assumptions for One-Way ANOVA:
    normality, equal variances, and independence.

    Parameters:
        groups (list of arrays): A list where each element is a group of data.
    """

    # Assumption 1: Normality check for each group using the Shapiro-Wilk test
    print("\nChecking normality using the Shapiro-Wilk test:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk Test: p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"  Group {i+1} is NOT normally distributed.")
        else:
            print(f"  Group {i+1} is approximately normally distributed.")

    # Assumption 2: Equal variances check using Levene's test
    print("\nChecking equal variances using Levene's test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test: p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  The variances are significantly different (violating equal variance assumption).")
    else:
        print("  The variances are approximately equal.")

    # Assumption 3: Independence is assumed if the data collection method was appropriate (random sampling)
    # This is not directly testable, but you should ensure the data was collected independently.
    print("\nAssumption of independence: Assuming that the data are independently sampled.")

    # Plot Q-Q plot for normality check
    print("\nVisualizing the Q-Q plot for normality check:")
    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i+1} Q-Q plot")
    plt.tight_layout()
    plt.show()

# Example: Simulating three groups of data
np.random.seed(0)

# Group 1: Normal distribution with mean=50, std=10
group_1 = np.random.normal(loc=50, scale=10, size=30)

# Group 2: Normal distribution with mean=55, std=10
group_2 = np.random.normal(loc=55, scale=10, size=30)

# Group 3: Normal distribution with mean=60, std=10
group_3 = np.random.normal(loc=60, scale=10, size=30)

# Checking assumptions for the groups
check_anova_assumptions([group_1, group_2, group_3])
```

Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Simulating data for two factors: Factor A (Treatment vs Control), Factor B (Time 1 vs Time 2)
np.random.seed(0)

# Factor A: Treatment (1) vs Control (0)
# Factor B: Time 1 (1) vs Time 2 (2)
n = 30  # Number of samples per group

# Treatment and Control groups for both Time 1 and Time 2
group_A1_B1 = np.random.normal(loc=50, scale=10, size=n)  # Treatment, Time 1
group_A1_B2 = np.random.normal(loc=55, scale=10, size=n)  # Treatment, Time 2
group_A2_B1 = np.random.normal(loc=45, scale=10, size=n)  # Control, Time 1
group_A2_B2 = np.random.normal(loc=50, scale=10, size=n)  # Control, Time 2

# Create a DataFrame for the data
data = pd.DataFrame({
    'Factor_A': ['Treatment']*n + ['Treatment']*n + ['Control']*n + ['Control']*n,
    'Factor_B': ['Time 1']*n + ['Time 2']*n + ['Time 1']*n + ['Time 2']*n,
    'Score': np.concatenate([group_A1_B1, group_A1_B2, group_A2_B1, group_A2_B2])
})

# Perform the Two-Way ANOVA
model = ols('Score ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the results
print(anova_table)

# Visualizing the interaction effect between Factor A and Factor B
plt.figure(figsize=(8, 6))
sns.pointplot(x='Factor_B', y='Score', hue='Factor_A', data=data, markers="o", linestyles='-', dodge=True)
plt.title('Interaction Plot for Two-Way ANOVA')
plt.xlabel('Factor B (Time)')
plt.ylabel('Scores')
plt.show()
```

 Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to plot F-distribution for different df1 and df2
def plot_f_distribution(df1_values, df2_values, x_range=(0, 5)):
    """
    Function to plot the F-distribution for different degrees of freedom (df1 and df2).

    Parameters:
    df1_values (list): List of numerator degrees of freedom (df1) to visualize.
    df2_values (list): List of denominator degrees of freedom (df2) to visualize.
    x_range (tuple): Range of x-axis values to plot the distribution over.
    """
    x = np.linspace(x_range[0], x_range[1], 500)

    plt.figure(figsize=(10, 6))

    # Plot F-distribution for different df1 and df2
    for df2 in df2_values:
        for df1 in df1_values:
            y = stats.f.pdf(x, df1, df2)
            plt.plot(x, y, label=f"df1={df1}, df2={df2}")

    plt.title('F-Distribution for Various Degrees of Freedom (df1, df2)', fontsize=14)
    plt.xlabel('F-statistic value', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend(title="Degrees of Freedom", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()

# Visualize the F-distribution for different df1 and df2 values
df1_values = [1, 2, 5]  # Numerator degrees of freedom
df2_values = [10, 20, 30]  # Denominator degrees of freedom

plot_f_distribution(df1_values, df2_values)
```

Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for three groups (Group 1, Group 2, Group 3)
np.random.seed(0)

# Group 1: Mean = 50, Std = 10
group_1 = np.random.normal(loc=50, scale=10, size=30)

# Group 2: Mean = 55, Std = 10
group_2 = np.random.normal(loc=55, scale=10, size=30)

# Group 3: Mean = 60, Std = 10
group_3 = np.random.normal(loc=60, scale=10, size=30)

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(group_1, group_2, group_3)

# Print the ANOVA result
print("ANOVA Test Results:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Visualization: Boxplots to compare the group means
data = [group_1, group_2, group_3]
group_labels = ['Group 1', 'Group 2', 'Group 3']

plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.xticks(ticks=np.arange(3), labels=group_labels)
plt.title('Boxplots of Group Means')
plt.ylabel('Values')
plt.show()

# Interpret the results
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from a normal distribution
np.random.seed(0)  # For reproducibility

# Parameters for the normal distribution
population_mean = 50  # Hypothesized population mean
population_std = 10   # Standard deviation
sample_size = 100     # Number of samples

# Generate random data
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Step 2: Perform One-Sample T-Test
t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)

# Step 3: Print the hypothesis test results
print("Hypothesis Testing Results:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Step 4: Visualizing the sample data
plt.figure(figsize=(8, 6))
plt.hist(sample_data, bins=20, color='skyblue', edgecolor='black')
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label='Hypothesized Mean')
plt.title('Histogram of Simulated Sample Data', fontsize=14)
plt.xlabel('Data Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()

# Step 5: Interpret the results
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(0)  # For reproducibility

# Parameters for the normal distribution
population_variance = 100  # Hypothesized population variance (sigma^2)
sample_size = 30           # Number of samples

# Generate random data with known variance (population variance)
sample_data = np.random.normal(loc=0, scale=np.sqrt(population_variance), size=sample_size)

# Step 2: Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)  # Using ddof=1 for sample variance

# Step 3: Perform Chi-square test for population variance
# Chi-square statistic: (n - 1) * sample_variance / hypothesized_variance
chi_square_stat = (sample_size - 1) * sample_variance / population_variance

# Step 4: Calculate p-value for two-tailed test
df = sample_size - 1  # Degrees of freedom (n - 1)
p_value_lower = stats.chi2.cdf(chi_square_stat, df)
p_value_upper = 1 - p_value_lower

# Two-tailed test: p-value is the sum of the two tail probabilities
p_value = p_value_lower + p_value_upper

# Step 5: Print the hypothesis test results
print("Hypothesis Testing Results:")
print(f"Chi-square Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Step 6: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data for two groups
# Group 1: Sample size = 1000, 200 successes
# Group 2: Sample size = 1000, 250 successes
n1 = 1000  # Sample size of Group 1
x1 = 200   # Number of successes in Group 1

n2 = 1000  # Sample size of Group 2
x2 = 250   # Number of successes in Group 2

# Step 2: Calculate sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Step 3: Calculate combined proportion (pooled proportion)
p_combined = (x1 + x2) / (n1 + n2)

# Step 4: Calculate Z-statistic for proportions
z_stat = (p1 - p2) / np.sqrt(p_combined * (1 - p_combined) * (1/n1 + 1/n2))

# Step 5: Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

# Step 6: Print the hypothesis test results
print("Hypothesis Testing Results:")
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Step 7: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate observed data
np.random.seed(0)

# Simulate 1000 data points with observed frequencies in 5 categories
observed = np.array([180, 210, 190, 210, 210])  # Observed frequencies for 5 categories

# Step 2: Calculate the expected frequencies (uniform distribution assumed)
expected = np.full_like(observed, 200)  # Expected frequency for each category is 200

# Step 3: Perform the Chi-square test for goodness of fit
chi2_stat, p_value = stats.chisquare(observed, expected)

# Step 4: Print the test results
print("Chi-Square Test Results:")
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Step 5: Visualize the observed and expected frequencies
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']

x = np.arange(len(categories))

plt.bar(x - 0.2, observed, 0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected, 0.4, label='Expected', color='orange')

plt.xticks(x, categories)
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.show()

# Step 6: Interpretation of Results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data follows the expected distribution.")
```
